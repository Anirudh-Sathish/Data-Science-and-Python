```python
# @ Author : Anirudh Sathish 
# @ Roll_no : CS20B1125 
# @ Author : Anirudh Sathish 
# @ Roll_no : CS20B1125 

# Importing the neccesary libraries 
import pandas as pd 
import numpy as np 
import sklearn
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.neighbors import NearestCentroid

# Reading the file 
df = pd.read_csv("diabetes.csv")

# Glucose values are zero in some cases , removing them 
for ind in df.index :
    if(df['Glucose'][ind] == 0):
        df = df.drop(ind)

# Now taking the columns that matter 
# After anlaysis Glucose seemed apt
glucose_d = df.iloc[:,[1]]

# This contains the output value
diabetes_output = df.iloc[:,[7]]

glucose_n = glucose_d.to_numpy()
# Normalizing the vector
normalized_glucose = glucose_n / np.sqrt(np.sum(glucose_n**2))

# Covertingback to dataframe 
glucose_d = pd.DataFrame(normalized_glucose, columns = ['Glucose'])

# Merging them all to a single dataframe
diabetes_related_data = pd.concat([glucose_d] , axis=1)

# Splitting the data into test and training dataset
X_train , X_test , Y_train , Y_test = train_test_split(diabetes_related_data ,diabetes_output , test_size=0.2 ,random_state = 42 )

def PredictionScore(prediction,Y_test):
    count = 0
    value = 0
    for ind in Y_test.index :
        if(prediction[count] == Y_test['Outcome'][ind] ):
            value+=1
        count+=1
    print("Accurate prediction is : ",(value/count)*100)
    
# Solution using (a) Minimum distance classifier 
model_min_dist_classifier = NearestCentroid()
# Training the classifier
model_min_dist_classifier.fit(X_train, Y_train.values.ravel())
min_dist_classifier_prediction = model_min_dist_classifier.predict(X_test)
print("Prediction For Minimum Distance Classifier")
print(min_dist_classifier_prediction)
PredictionScore(min_dist_classifier_prediction,Y_test)

# Solution using (b) Kmeans Algorithm 
kmeans_cluster_classification = KMeans(n_clusters = 2, random_state=42)

# Training the model 
kmeans_model = kmeans_cluster_classification.fit(X_train)

# Predicted value for test 
kmeans_prediction = kmeans_model.predict(X_test)

print("Prediction using K means : ", kmeans_prediction)

PredictionScore(kmeans_prediction,Y_test)
```
